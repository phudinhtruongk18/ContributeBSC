```python
import pandas as pd
```

```python
url = 'https://raw.githubusercontent.com/phudinhtruongk18/test_data/main/data.csv'
```

```python
df = pd.read_csv(url)
```

```python
columns = ["Txhash", "Value_IN(BNB)","From","To","ContractAddress", "Status", "ErrCode", "Method"]
new_df = df[columns]
```

```python
filtered_df = new_df[new_df['Method'] == 'Contribute']
```

```python
filtered_df["Status"].unique()
```

```python
filtered_df = filtered_df[filtered_df["ErrCode"].isna() | filtered_df["Status"].isna()]
```

```python
filtered_df = filtered_df.reset_index(drop=True)
```

```python
filtered_df.head(10)
```

```python
filtered_df.count()
```

```python
filtered_df["From"].nunique()
```

```python
filtered_df.head()
```

```python
# Create the "extracted" DataFrame
extracted = filtered_df.groupby("From").agg(
    count=("Value_IN(BNB)", "count"),
    sum=("Value_IN(BNB)", "sum")
    ).reset_index()
```

```python
extracted["url"] = extracted["From"].apply(lambda x: f"https://bscscan.com/address/{x}")
```

```python
extracted = extracted.sort_values(by="sum")
```

```python
extracted.head(100)
```

```python
desired_value = "0x005055e06c5e0c92c94fbf89bd5a5eeae387dda5"
filtered_rows = filtered_df[filtered_df["From"] == desired_value]
filtered_rows
```

```python
# Split into two DataFrames based on condition
df_sum_lt_1 = extracted[extracted['sum'] < 1]
df_sum_gte_1 = extracted[extracted['sum'] >= 1]

# Calculate sum for each DataFrame
big_sum_lt_1 = df_sum_lt_1['sum'].sum()
big_sum_gte_1 = df_sum_gte_1['sum'].sum()

# Print the two big sums
print("Big Sum for sum < 1:", big_sum_lt_1)
print("Big Sum for sum >= 1:", big_sum_gte_1)
```
